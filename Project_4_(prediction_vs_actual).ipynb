```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
subang = pd.read_excel('OA GRU.xlsx', sheet_name = 'Subang')
```

```python
subang
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(subang['PSO-GRU'], color = 'green', label = 'PSO-GRU-1')
ax.plot(subang['WOA-GRU'], color = 'red', label = 'WOA-GRU-1')
ax.plot(subang['MFO-GRU'], color = 'blue', label = 'MFO-GRU-1')
ax.plot(subang['Actual ET0'], color = 'black', linestyle='--', label = 'Actual ' + r'ET$_{0}$')
ax.legend(bbox_to_anchor=(1, 1), fontsize=20)
plt.xticks([0,1,2,3,4,5], labels=['2016-09', '2016-10', '2016-11', '2016-12', '2017-01', '2017-02'], fontsize=20) 
plt.xlabel('Month', fontsize=22)
plt.ylabel(r'ET$_{0}$' +' (mm/day)', fontsize=22)
plt.yticks(fontsize=20)
plt.title('Subang', fontsize=25)

plt.savefig('Subang' +'.png', bbox_inches='tight')
```

```python
# Residual plot
subang['diff_pso1'] = subang['Actual ET0'] -  subang['PSO-GRU']
subang['diff_woa1'] = subang['Actual ET0'] -  subang['WOA-GRU']
subang['diff_mfo1'] = subang['Actual ET0'] -  subang['MFO-GRU']
subang_new = subang[['diff_pso1', 'diff_woa1', 'diff_mfo1']].reset_index()
subang_melt = subang_new.melt(id_vars ='index', var_name='diff',  value_name='Residuals')

f, ax = plt.subplots(figsize=(10, 6))

palette = {'diff_pso1':'green', 'diff_woa1':'red', 'diff_mfo1':'blue' }
markers = {'diff_pso1':'o','diff_woa1':'o', 'diff_mfo1':'o'}
sns.scatterplot(x='index', y="Residuals", hue='diff', data=subang_melt, s=80, palette=palette, style='diff', markers=markers)
L = ax.legend(loc = 'lower right', fontsize=20)
L.get_texts()[0].set_text('PSO-GRU-1')
L.get_texts()[1].set_text('WOA-GRU-1')
L.get_texts()[2].set_text('MFO-GRU-1')
plt.yticks(fontsize=20)
plt.ylabel('Residuals', fontsize=22)
plt.title('Subang', fontsize=25)
plt.xticks([0,1,2,3,4,5], labels=" ") # remove xtick label
plt.tick_params(bottom = False) # remove xticks
plt.xlabel('')
```

```python
subang_metrics = pd.read_excel('OA GRU.xlsx', sheet_name = 'Subang1', header=None, index_col=0)
ipoh_metrics = pd.read_excel('OA GRU.xlsx', sheet_name = 'Ipoh1', header=None, index_col=0)
mudzam_metrics = pd.read_excel('OA GRU.xlsx', sheet_name = 'Muadzam1', header=None, index_col=0)
langkawi_metrics = pd.read_excel('OA GRU.xlsx', sheet_name = 'Langkawi1', header=None, index_col=0)
subang_metrics
```

# Prepare the dataframes for MAE, RMSE, MAPE and KGE

```python
mae = []
for i in langkawi_metrics, ipoh_metrics, subang_metrics, mudzam_metrics:
  mae.append(list(i.iloc[0,:]))

rmse = []
for i in langkawi_metrics, ipoh_metrics, subang_metrics, mudzam_metrics:
  rmse.append(list(i.iloc[1,:]))
  
mape = []
for i in langkawi_metrics, ipoh_metrics, subang_metrics, mudzam_metrics:
  mape.append(list(i.iloc[2,:]))

kge = []
for i in langkawi_metrics, ipoh_metrics, subang_metrics, mudzam_metrics:
  kge.append(list(i.iloc[3,:]))  
```

```python
mape
```

```python
stations = ['Langkawi', 'Ipoh', 'Subang', 'Muadzam']
cols = ['PSO-GRU-1','WOA-GRU-1','MFO-GRU-1','PSO-GRU-2', 'WOA-GRU-2', 'MFO-GRU-2']
mae_df = pd.DataFrame(mae, stations, cols)
rmse_df = pd.DataFrame(rmse, stations, cols)
mape_df = pd.DataFrame(mape, stations, cols)
kge_df = pd.DataFrame(kge, stations, cols)
```

```python
mae_df
```

```python
N = 4
ind = np.arange(N) 
width = 0.1
plt.figure(figsize=(30, 40))

# MAE
plt.subplot(4, 1, 1)
bar1 = plt.bar(ind, mae_df['PSO-GRU-1'], width, color='tab:green')
bar2 = plt.bar(ind+width, mae_df['WOA-GRU-1'], width, color='tab:red')
bar3 = plt.bar(ind+width*2, mae_df['MFO-GRU-1'], width, color='tab:blue')
bar4 = plt.bar(ind+width*3, mae_df['PSO-GRU-2'], width, fill=False, hatch='*', edgecolor='tab:green')
bar5 = plt.bar(ind+width*4, mae_df['WOA-GRU-2'], width, fill=False, hatch='*', edgecolor='tab:red' )
bar6 = plt.bar(ind+width*5, mae_df['MFO-GRU-2'], width, fill=False, hatch='*', edgecolor='tab:blue')
plt.legend(['PSO-GRU-1','WOA-GRU-1','MFO-GRU-1','PSO-GRU-2', 'WOA-GRU-2', 'MFO-GRU-2'],
           bbox_to_anchor=(1.0, 1.0), prop={'size': 60}, fontsize=55)
plt.yticks(fontsize=52)
plt.xticks([0.25,1.25,2.25, 3.25], labels=['Pulau Langkawi', 'Ipoh', 'Subang', 'Muadzam Shah'], fontsize=50)
plt.title('MAE (mm/day)', fontsize= 70)

# RMSE
plt.subplot(4, 1, 2)
bar1 = plt.bar(ind, rmse_df['PSO-GRU-1'], width, color='tab:green')
bar2 = plt.bar(ind+width, rmse_df['WOA-GRU-1'], width, color='tab:red')
bar3 = plt.bar(ind+width*2, rmse_df['MFO-GRU-1'], width, color='tab:blue')
bar4 = plt.bar(ind+width*3, rmse_df['PSO-GRU-2'], width, fill=False, hatch='*', edgecolor='tab:green')
bar5 = plt.bar(ind+width*4, rmse_df['WOA-GRU-2'], width, fill=False, hatch='*', edgecolor='tab:red' )
bar6 = plt.bar(ind+width*5, rmse_df['MFO-GRU-2'], width, fill=False, hatch='*', edgecolor='tab:blue')
plt.yticks(fontsize=52)
plt.xticks([0.25,1.25,2.25, 3.25], labels=['Pulau Langkawi', 'Ipoh', 'Subang', 'Muadzam Shah'], fontsize=50)
plt.title('RMSE (mm/day)', fontsize= 70)

# MAPE
plt.subplot(4, 1, 3)
bar1 = plt.bar(ind, mape_df['PSO-GRU-1'], width, color='tab:green')
bar2 = plt.bar(ind+width, mape_df['WOA-GRU-1'], width, color='tab:red')
bar3 = plt.bar(ind+width*2, mape_df['MFO-GRU-1'], width, color='tab:blue')
bar4 = plt.bar(ind+width*3, mape_df['PSO-GRU-2'], width, fill=False, hatch='*', edgecolor='tab:green')
bar5 = plt.bar(ind+width*4, mape_df['WOA-GRU-2'], width, fill=False, hatch='*', edgecolor='tab:red' )
bar6 = plt.bar(ind+width*5, mape_df['MFO-GRU-2'], width, fill=False, hatch='*', edgecolor='tab:blue')
plt.yticks(fontsize=52)
plt.xticks([0.25,1.25,2.25, 3.25], labels=['Pulau Langkawi', 'Ipoh', 'Subang', 'Muadzam Shah'], fontsize=50)
plt.title('MAPE (%)', fontsize= 70)

# KGE
plt.subplot(4, 1, 4)
bar1 = plt.bar(ind, kge_df['PSO-GRU-1'], width, color='tab:green')
bar2 = plt.bar(ind+width, kge_df['WOA-GRU-1'], width, color='tab:red')
bar3 = plt.bar(ind+width*2, kge_df['MFO-GRU-1'], width, color='tab:blue')
bar4 = plt.bar(ind+width*3, kge_df['PSO-GRU-2'], width, fill=False, hatch='*', edgecolor='tab:green')
bar5 = plt.bar(ind+width*4, kge_df['WOA-GRU-2'], width, fill=False, hatch='*', edgecolor='tab:red' )
bar6 = plt.bar(ind+width*5, kge_df['MFO-GRU-2'], width, fill=False, hatch='*', edgecolor='tab:blue')
plt.yticks(fontsize=52)
plt.xticks([0.25,1.25,2.25, 3.25], labels=['Pulau Langkawi', 'Ipoh', 'Subang', 'Muadzam Shah'], fontsize=50)
plt.title('KGE', fontsize= 70)

plt.subplots_adjust(bottom=0.1, 
                    top=0.9,  
                    hspace=0.5)

plt.savefig('Performance Metrics' +'.png', bbox_inches='tight')
```
